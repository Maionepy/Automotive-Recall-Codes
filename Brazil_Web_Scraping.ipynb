```python
from selenium import webdriver
import pandas as pd
chromeOptions = webdriver.ChromeOptions()
chromeOptions.add_experimental_option('useAutomationExtension', False)

df = pd.DataFrame(columns=['Título','Protocolo','Data Início','Descrição','Medida de Correção','Relato de Acidentes','Componentes','Data de Constatação','Defeito','Tipo de Risco','Descrição do Risco','Implicações do Risco','Setor','Tipo de Produto','Fabricante','País de Origem','País de Exportação','Modelo','Ano de Fabricação','Quantidade Afetada','Link','PDF'])
driver = webdriver.Chrome(options=chromeOptions, desired_capabilities=chromeOptions.to_capabilities(),executable_path='chromedriver.exe')

elementos_existentes = 0

for i in range(2500):
    
    titulo = None
    protocolo = None
    data_inicio = None
    descricao = None
    medida_de_correcao = None
    relato_de_acidentes = None
    
    componente = None
    data_constatacao = None
    defeito = None
    tipo_de_risco = None
    descricao_do_risco = None
    implicacoes_do_risco = None
    
    setor = None
    tipo_de_produto = None
    fabricante = None
    pais_de_origem = None
    pais_de_exportacao = None
    modelo = None
    ano = None
    qtde = None
    
    
    try:
        link = 'http://portal.mj.gov.br/recall/principal/detailRecall/' + str(i)
        driver.get(link)
        titulo = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[1]/table/tbody/tr[1]/td[2]').text
        protocolo = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[1]/table/tbody/tr[2]/td[2]').text
        data_inicio = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[1]/table/tbody/tr[3]/td[2]').text
        descricao = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[1]/table/tbody/tr[4]/td[2]').text
        medida_de_correcao = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[1]/table/tbody/tr[5]/td[2]').text
        relato_de_acidentes = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[1]/table/tbody/tr[6]/td[2]').text
        
        elementos_existentes += 1
        driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/ul/li[2]/a').click()
        componente = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[2]/table/tbody/tr[1]/td[2]').text
        data_constatacao = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[2]/table/tbody/tr[2]/td[2]').text
        defeito = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[2]/table/tbody/tr[3]/td[2]').text
        tipo_de_risco = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[2]/table/tbody/tr[4]/td[2]').text
        descricao_do_risco = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[2]/table/tbody/tr[5]/td[2]').text
        implicacoes_do_risco = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[2]/table/tbody/tr[6]/td[2]').text
        
        driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/ul/li[3]/a').click()
        
        try:
            setor = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[3]/table/tbody/tr[2]/td[2]').text
            tipo_de_produto = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[3]/table/tbody/tr[3]/td[2]').text
            fabricante = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[3]/table/tbody/tr[4]/td[2]').text
            pais_de_origem = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[3]/table/tbody/tr[5]/td[2]').text
            pais_de_exportacao = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[3]/table/tbody/tr[6]/td[2]').text
            modelo = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[3]/table/tbody/tr[7]/td[2]').text
            ano = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[3]/table/tbody/tr[8]/td[2]').text
            qtde = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[3]/table/tbody/tr[10]/td[1]/span').text
        except:
            pass
        
        driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/ul/li[5]/a').click()
        try:
            file = driver.find_element_by_xpath('/html/body/div[2]/main/div/div/div[4]/section/div[3]/div[5]/table/tbody/tr/td[2]/a')
            filename = file.text
            file.click()
        except:
            pass
        
        df_aux = pd.DataFrame([[titulo, protocolo,data_inicio,descricao,medida_de_correcao,relato_de_acidentes,componente,data_constatacao,defeito,tipo_de_risco,descricao_do_risco,implicacoes_do_risco,setor,tipo_de_produto,fabricante,pais_de_origem,pais_de_exportacao,modelo,ano,qtde,link,filename]], columns=['Título','Protocolo','Data Início','Descrição','Medida de Correção','Relato de Acidentes','Componentes','Data de Constatação','Defeito','Tipo de Risco','Descrição do Risco','Implicações do Risco','Setor','Tipo de Produto','Fabricante','País de Origem','País de Exportação','Modelo','Ano de Fabricação','Quantidade Afetada','Link','PDF'])
        df = pd.concat([df,df_aux],ignore_index=True)
        if i==501 or i==1000 or i==1499 or i==1969:
            df.to_excel('Base_Brazil_Senacon_General.xlsx',index=False)
            driver.quit()
            driver= webdriver.Chrome(options=chromeOptions, desired_capabilities=chromeOptions.to_capabilities(),executable_path='chromedriver.exe')

        
    except Exception as e:
        pass
    
df.to_excel('Base_Brazil_Senacon_General.xlsx',index=False)
```
